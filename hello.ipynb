## Etape 3 - Affichage de variables

```java
// Déclaration et initialisation
        String nomProduit1 = "Stylo";
        double prixUnitaire1 = 1.20;
        int quantite1 = 5;

        String nomProduit2 = "Cahier";
        double prixUnitaire2 = 3.50;
        int quantite2 = 2;

        double total = (prixUnitaire1 * quantite1) + (prixUnitaire2 * quantite2);

        // Affichage formaté
        // %-10s : Texte aligné à gauche sur 10 caractères
        // %6.2f : Nombre décimal sur 6 caractères avec 2 chiffres après la virgule
        // %3d   : Entier sur 3 caractères
        
        System.out.println("+------------+--------+-----+--------+");
        System.out.printf("| %-10s | %-6s | %-3s | %-6s |%n", "Produit", "PU", "Qté", "Prix");
        System.out.println("+------------+--------+-----+--------+");
        
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit1, prixUnitaire1, quantite1, (prixUnitaire1 * quantite1));
        
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit2, prixUnitaire2, quantite2, (prixUnitaire2 * quantite2));
        
        System.out.println("+------------+--------+-----+--------+");
        System.out.printf("| TOTAL                     | %6.2f |%n", total);
        System.out.println("+------------+--------+-----+--------+");
```
