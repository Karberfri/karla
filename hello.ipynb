## Etape 3 - Affichage de variables

```java
// Déclaration et initialisation
        String nomProduit1 = "Stylo";
        double prixUnitaire1 = 1.20;
        int quantite1 = 5;

        String nomProduit2 = "Cahier";
        double prixUnitaire2 = 3.50;
        int quantite2 = 2;

        double total = (prixUnitaire1 * quantite1) + (prixUnitaire2 * quantite2);

        // Affichage formaté
        // %-10s : Texte aligné à gauche sur 10 caractères
        // %6.2f : Nombre décimal sur 6 caractères avec 2 chiffres après la virgule
        // %3d   : Entier sur 3 caractères
        
        System.out.println("+------------+--------+-----+--------+");
        System.out.printf("| %-10s | %-6s | %-3s | %-6s |%n", "Produit", "PU", "Qté", "Prix");
        System.out.println("+------------+--------+-----+--------+");
        
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit1, prixUnitaire1, quantite1, (prixUnitaire1 * quantite1));
        
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit2, prixUnitaire2, quantite2, (prixUnitaire2 * quantite2));
        
        System.out.println("+------------+--------+-----+--------+");
        System.out.printf("| TOTAL                     | %6.2f |%n", total);
        System.out.println("+------------+--------+-----+--------+");
```

## Etape 4 - Saisie des informations

```java
// 1. Initialisation du scanner pour la saisie
        Scanner clavier = new Scanner(System.in);

        // 2. Déclaration des variables (Étape 3)
        String nomProduit1 = "Stylo";
        double prixUnitaire1 = 1.20;
        
        String nomProduit2 = "Cahier";
        double prixUnitaire2 = 3.50;

        // 3. Saisie des quantités (Étape 4)
        System.out.print("Entrez la quantité de " + nomProduit1 + " : ");
        int quantite1 = clavier.nextInt();

        System.out.print("Entrez la quantité de " + nomProduit2 + " : ");
        int quantite2 = clavier.nextInt();

        // 4. Calculs
        double total1 = prixUnitaire1 * quantite1;
        double total2 = prixUnitaire2 * quantite2;
        double totalGeneral = total1 + total2;

        // 5. Affichage du tableau formaté
System.out.println("+----------+--------+-----+--------+");
        System.out.println("| Produit  | PU (€) | Qté | Prix € |");
        System.out.println("+----------+--------+-----+--------+");

        System.out.println("| Stylo    | 1.20   |  5  |  6.00  |");
        System.out.println("| Cahier   | 3.50   |  2  |  7.00  |");

        System.out.println("+----------+--------+-----+--------+");
        System.out.println("| TOTAL                    | 13.00 |");
        System.out.println("+----------+--------+-----+--------+");
        
        // Ligne produit 1
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit1, prixUnitaire1, quantite1, total1);
        
        // Ligne produit 2
        System.out.printf("| %-10s | %6.2f | %3d | %6.2f |%n", 
                          nomProduit2, prixUnitaire2, quantite2, total2);
        
        System.out.println("+------------+--------+-----+--------+");
        
        // Ligne total
        System.out.printf("| TOTAL                     | %6.2f |%n", totalGeneral);
        System.out.println("+------------+--------+-----+--------+");
```
